```python
#Import required libraries to plot the graph
import matplotlib.pyplot as plt

#We have an example dictionary like this
dict={"Average Energy 0.1": 120.3, "Average Energy 0.2": 133.3,"Average Energy 0.3": 83.3,"Average Energy 0.4": 150.3,"Windspeed 0.1": 20.3, "Windspeed 0.2": 125.3,"Windspeed 0.3": 60.3}

print(dict)

#Todo: Plot a graph of Average Energy
#Create a new dictionary to store only Average Energy key value pairs
newdict={}

#Store the key value pairs which have the term 'Average Energy' in their key to newdict
for key in dict.keys():
    if "Average Energy" in key:
        newdict[key] = dict[key]
        
print(newdict)

#x axis values will be the dictionary keys
x=newdict.keys()
print(x)
# corresponding y axis values will be the values
y=[]
for key in newdict.keys():
    y.append(newdict[key])

print(y)
  
# plotting the points 
plt.plot(x, y)
  
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
  
# giving a title to my graph
plt.title('Average Energy Graph!')
  
# function to show the plot
plt.show()

#Todo: Plot a graph of Windspeed
#Create a new dictionary to store only Windspeed key value pairs
newdict={}

#Store the key value pairs which have the term 'Windspeed' in their key to newdict
for key in dict.keys():
    if "Windspeed" in key:
        newdict[key] = dict[key]
        
print(newdict)

#x axis values will be the dictionary keys
x=newdict.keys()
print(x)
# corresponding y axis values will be the values
y=[]
for key in newdict.keys():
    y.append(newdict[key])

print(y)
  
# plotting the points 
plt.plot(x, y)
  
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
  
# giving a title to my graph
plt.title('Windspeed Graph!')
  
# function to show the plot
plt.show()
```

```python
#Export the entire dictionary as a CSV file
import csv

dict={"Average Energy 0.1": 120.3, "Average Energy 0.2": 133.3,"Average Energy 0.3": 83.3,"Average Energy 0.4": 150.3,"Windspeed 0.1": 20.3, "Windspeed 0.2": 125.3,"Windspeed 0.3": 60.3}

with open('output.csv', 'w') as output:
    writer = csv.writer(output)
    writer.writerow(["Key","Value"])
    for key, value in dict.items():
        writer.writerow([key, value]) 
        
#If you want output in .xlsx(excel) format just open the .csv file in excel and save as .xlsx
```
